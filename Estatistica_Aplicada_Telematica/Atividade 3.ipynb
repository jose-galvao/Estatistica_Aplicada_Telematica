### <center> Questão 1 </center>

**A lista** é um modo nativo de o python guardar varios valores, informações ou dados em uma unica variavel. Na hora de imprimir essa váriavel dá para você selecionar o indice, caso não queira dar print na lista interira exemplo:

```python
lista = ['amarelo', 9, 3.6, b'azul']
print(lista[0])
```

Nesse caso, pedi para imprimir o **indice zero da variável lista** que é o primeiro dado (amarelo) que aparece dentro da lista.

**O dicionario** usa uma chave para acessar certo valor ou valores. É como se fosse uma lista dentro de uma lista e para acessar certo dado, você terá que colocar dois índices. O indice da "página" que seria o nome que você deu a pagina chave e o indice especifico que é o que está dentro da chave de acesso. Exemplo:

```python
dici = {'pagina1': ['Arroz', 'Abacate', 'Abobora'],
             'pagina2':['Baleia', 'brigadeiro','bacia']}
print(dici['pagina2'][1])
```

**O comando serie do pandas** parece com um dataframe, porém é uma lista com uma unica coluna de dados e uma coluna com os indices dos dados. Por exemplo:

```python
import pandas
serie = pandas.Series(lista)
print(serie)
```

Usei a mesma lista do exemplo de listas e ele criou uma especie de table para os dados.

**O array Numpy** vai também usar a lista como exemplo para ver a diferença.

```python
import numpy
array = numpy.array(lista)
print(array)
```

Ao contrário do series panda o Array numpy não imprime os dados como uma tabela na vertical e nem coloca indices nos dados.
